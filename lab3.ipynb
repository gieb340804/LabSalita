4.2 Вычислить вектор средних значений времени работы и среднеквадратичных отклонений времени работы для каждой длины сортируемой последовательности;

```python
from random import randint
import timeit
import numpy as np

def generator(n):
    data = []
    for i in range(n):
        data.append(randint(0,n))

    return data    

def permutationSort(massiv):
    n = len(massiv)
    
    while True:
        swapped = False
        for i in range(n - 1):
            if massiv[i] > massiv[i+1]:
                temp = massiv[i]
                massiv[i] = massiv[i+1]
                massiv[i+1] = temp
                swapped = True
        if swapped == False:
            break
    return massiv



if __name__ == "__main__":
    result = []
    for i in range(1000, 11000, 1000):
        print(i)
        # data = generator(i)
        # print(data)
        
        timings = []
        for s in range(10):
            data = generator(i)
            timerSort = timeit.timeit(lambda: permutationSort(data), number= 1)
            timings.append(timerSort)
            print(timings)
        result.append(timings)
            # print(f'{timerSort} секунд')
        # print(data) 
            
        print(result)
            
    numpy_array = np.array(result)
    print(numpy_array)
    mean_times = np.mean(numpy_array, axis=1)
    std_devs = np.std(numpy_array, axis=1)
    print("Средние значения времени работы:")
    print(mean_times)
    print("Среднеквадратичные отклонения времени работы:")
    print(std_devs)
```
